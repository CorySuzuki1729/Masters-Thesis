```python
# # Extrapolating data to get future predictions. 

# def evolve_pde(u0, dx, dy, dt, steps):
#     """
#     Evolves the PDE u_t = 0.0249u - 0.0973u_x + 0.0104u_y
#     using finite difference for given time steps.
    
#     Parameters:
#         u0: np.ndarray, initial image (2D array)
#         dx, dy: spatial resolution
#         dt: time step
#         steps: number of future steps to predict
        
#     Returns:
#         A list of predicted images.
#     """
#     u = u0.copy()
#     nx, ny = u.shape
#     images = [u.copy()]
    
#     for _ in range(steps):
#         u_x = (np.roll(u, -1, axis=0) - np.roll(u, 1, axis=0)) / (2 * dx)
#         u_y = (np.roll(u, -1, axis=1) - np.roll(u, 1, axis=1)) / (2 * dy)
        
#         # PDE update
#         u_new = u + dt * (0.0249 * u - 0.0973 * u_x + 0.0104 * u_y)
#         images.append(u_new.copy())
#         u = u_new
    
#     return images

def evolve_pde_with_sampling(u0, dx, dy, dt, start_time, end_time, sample_times):
    """
    Evolve PDE and return images at specified times.
    
    Parameters:
        u0: np.ndarray, initial image at start_time
        dx, dy: spatial resolution
        dt: simulation time step
        start_time: starting time (e.g., 50.0)
        end_time: ending time (e.g., 55.0)
        sample_times: list of times to save images at (e.g., [51.25, 51.5, ..., 55.0])
        
    Returns:
        Dictionary mapping sampled times to image arrays.
    """
    u = u0.copy()
    t = start_time
    sampled_images = {}
    sample_times = sorted(sample_times)
    next_sample_idx = 0
    
    while t < end_time + dt/2:
        if abs(t - sample_times[next_sample_idx]) < dt/2:
            sampled_images[round(t, 4)] = u.copy()
            next_sample_idx += 1
            if next_sample_idx >= len(sample_times):
                break
        
        # Compute spatial derivatives
        u_x = (np.roll(u, -1, axis=0) - np.roll(u, 1, axis=0)) / (2 * dx)
        u_y = (np.roll(u, -1, axis=1) - np.roll(u, 1, axis=1)) / (2 * dy)
        
        # Update PDE
        u_new = u + dt * (0.0249 * u - 0.0973 * u_x + 0.0104 * u_y)
        u = u_new
        t += dt

    return sampled_images

```

```python
def forward_euler_pde(u0, dx, dy, dt, steps):
    """
    Evolves the PDE using a forward Euler scheme over fixed integer steps.
    
    Parameters:
        u0: np.ndarray, initial image at Image_50
        dx, dy: spatial resolution
        dt: time step size (should be 1.0 if one step = one image)
        steps: number of future images to generate (e.g., 5 for Image_51 to Image_55)
        
    Returns:
        List of images [Image_50, Image_51, ..., Image_55]
    """
    u = u0.copy()
    images = [u.copy()]
    
    for _ in range(steps):
        u_x = (np.roll(u, -1, axis=0) - np.roll(u, 1, axis=0)) / (2 * dx)
        u_y = (np.roll(u, -1, axis=1) - np.roll(u, 1, axis=1)) / (2 * dy)
        
        u_t = 0.0249 * u - 0.0973 * u_x + 0.0104 * u_y
        u = u + dt * u_t
        images.append(u.copy())
    
    return images
```

```python
# Assume image_50 is your initial condition
dx = dy = 1.0     # spatial step size
dt = 1.0          # since you're stepping once per image
steps = 5         # Image_51 to Image_55

predicted_images = forward_euler_pde(image_45, dx, dy, dt, steps)

# To view Image_53:
import matplotlib.pyplot as plt
plt.imshow(predicted_images[5], cmap='gray')  # Index 3 = Image_53
plt.title("Predicted Image 50")
plt.show()
```
